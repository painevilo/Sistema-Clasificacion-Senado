```python
# Generamos las variables Independientes (PRICES) y las independientes (FEATURES)
prices = df2['MEDV']
features = df2.drop('MEDV', axis = 1)
#
#Para probar con las de mayor correlaci√≥n con PRICES : RM,LSTAT,PTRATIO
#
#features = df2.drop(['CRIM','ZN','INDUS','CHAS','NOX','AGE','DIS','RAD','TAX','B','MEDV'], axis=1)
#
#Para probar s√≥lo con RM
#
#features = df2.drop(['CRIM','ZN','INDUS','CHAS','NOX','AGE','DIS','RAD','TAX','B','MEDV','PTRATIO','LSTAT'], axis=1)

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state =1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Tipo de modelo
modelo_x = LinearRegression()
# Entrenamiento del modelo
modelo_x.fit(X_train, y_train)
# Validaci√≥n del modelo
y_pred = modelo_x.predict(X_test)
# Evaluaci√≥n del modelo
print("R2 :%.2f" % r2_score(y_test, y_pred))
print("MSE: %.2f" % mean_squared_error(y_pred, y_test))
print("Score del modelo : %.2f" % modelo_x.score(X_test,y_test))
```

```python
# Guardamos la prediccion original
#
y_pred_original=y_pred
```

```python
# Incrementamos CRIM en un 1%
#
features2 = df2.drop('MEDV', axis = 1)
features2['CRIM'] = features2['CRIM']*1.01
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features2, prices, test_size=0.2, random_state =1)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Tipo de modelo
modelo_x = LinearRegression()
# Entrenamiento del modelo
modelo_x.fit(X_train, y_train)
# Validaci√≥n del modelo
y_pred = modelo_x.predict(X_test)
# Evaluaci√≥n del modelo
print("R2 :%.2f" % r2_score(y_test, y_pred))
print("MSE: %.2f" % mean_squared_error(y_pred, y_test))
print("Score del modelo : %.2f" % modelo_x.score(X_test,y_test))
```
